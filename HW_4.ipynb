```python
sns.histplot(df['yr_built'], bins=50)
plt.title('Распределение недвижимости по годам')
plt.xlabel('Год постройки')
plt.ylabel('Кол-во объектов, шт.');
```

```python
# Готовим данные для графика в виде сводной таблицы
data = df.groupby('yr_built')['price'].sum().reset_index()
data.head()
```

```python
plt.figure(figsize=(6, 4))

plt.scatter(data['yr_built'], data['price'])

plt.title('График разброса стоимости объекта по годам')
plt.xlabel('Год постройки')
plt.ylabel('Стоимость объекта');
```

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.<br>
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df[['price','bedrooms','bathrooms','floors','view','condition','yr_built']].corr()
corr_matrix = np.round(corr_matrix, 2)
corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=3.5, cmap='coolwarm')
plt.title('Correlation matrix');
```
