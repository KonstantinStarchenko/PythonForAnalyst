* 4.1 Сколько в среднем стоят дома, у которых 2 спальни?

```python
df_2_bathrooms=df[(df['bathrooms']==2)][['price']].mean()
df_2_bathrooms
```

* 4.2 Какая в среднем общая площадь домов, у которых стоимость больше 600 000?

```python
df_sqft_lot=df[(df['price']>600000)][['sqft_lot']].mean()
df_sqft_lot
```

* 4.3 Как много домов коснулся ремонт?

```python
df_renovated=df[(df['yr_renovated']>0)][['yr_renovated']].shape[0]
df_renovated
```

* 4.4 Насколько в среднем стоимость домов с оценкой grade <br>
домов выше 10 отличается от стоимости домов с оценкой grade
меньше 4?

```python
df_grade1=df[(df['grade']>10)][['price']].mean()
df_grade1
```

```python
df_grade2=df[(df['grade']<4)][['price']].mean()
df_grade2
```
